```python
import os
import pandas as pd 
```

```python
files = os.listdir("../data")
files
```

```python
dfs = [] 

for file in files:
    df_temp = pd.read_csv(f"../data/{file}")
    dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['product'].unique()
```

```python
df["price"] = df["price"].str.replace("$", "", regex=False).astype(float)
```

```python
df = df[df["product"] == "pink morsel"].copy()
```

```python
df["sales"] = df["price"] * df["quantity"]
```

```python
df = df[["sales", "date", "region"]]
```
