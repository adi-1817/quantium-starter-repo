## Converting date column to appropriate data type

In our data, 'date' is an object data type. 
We convert it to datetime64 data type for better analysis

```python
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

## GROUPING BY & SORTING 

1. Grouping by date and then performing a summation on sales column will give us the exact sale figure happened on a particular day. 
2. As this is a time series data, we then sort the values by date

```python
df = df.groupby("date")["sales"].sum().reset_index()
df = df.sort_values("date")
df.head(10)
```

## Visualization & Answering Business Problem

1. A line chart is plotted with 'date' as x-axis, 'sales' as y-axis in order to understand the trends over a particular time.
2. A vertical line is added on the exact business date requirement which is 15th of Jan, 2021 answering the question.

```python
fig = px.line(df, x = "date", y="sales", title = "Sales Over Time")
fig.add_vline(x = "2021-01-15", line_dash = "dash", line_color = "red")
fig.show()
```

# Were sales higher before or after the Pink Morsel price increase on the 15th of January, 2021?

Answer: The sales grew after the pink morsel price increase as the chart clearly depicts.
